```python
# 1. 基本数值不等式判断

a, b = 10, 20

print(a < b)   # True
print(a > b)   # False
print(a <= b)  # True
print(a >= b)  # False
print(a == b)  # False
print(a != b)  # True
```

```python
# 2. 结合 if 语句进行不等式判断

x = 15

if x > 10:
    print("x 大于 10")
elif x < 10:
    print("x 小于 10")
else:
    print("x 等于 10")
```

```python
# 3. 含绝对值不等式

from sympy import symbols, Abs, solve_univariate_inequality

#示例 1：求解 |x - 3| < 5
x = symbols('x')

inequality = Abs(x - 3) < 5
solution = solve_univariate_inequality(inequality, x)

print("解集1：", solution)  # 输出 (-2, 8)

# 示例 2：求解 |2x + 1| ≥ 3
inequality = Abs(2*x + 1) >= 3
solution = solve_univariate_inequality(inequality, x)

print("解集2：", solution)  # 输出 (-∞, -2] ∪ [1, ∞)
```

```python
# 4.一元二次不等式

from sympy import solve

# 示例 3：求解x**2 - 5x +6<=0 
inequality = (x**2 - 5*x + 6) <= 0
solution = solve_univariate_inequality(inequality, x)

print("解集3：", solution)  # 输出 [2, 3]

# 示例4：x^2 + 4x + 5 > 0
inequality = x**2 + 4*x + 5 > 0
solution = solve_univariate_inequality(inequality, x)

print("解集4：", solution)  # 输出 (-∞, ∞)  (恒成立)
```

```python
# 5. 求解线性不等式组

import numpy as np
import matplotlib.pyplot as plt

# 定义 x 范围
x = np.linspace(-10, 10, 400)
y1 = (4 - x) / 2  # 由 x + 2y = 4 变形得 y = (4 - x) / 2
y2 = (3*x - 2)  # 由 3x - y = 2 变形得 y = 3x - 2

# 绘制不等式区域
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label=r'$x + 2y \geq 4$', color='r')
plt.plot(x, y2, label=r'$3x - y \leq 2$', color='b')

# 填充可行解区域
y_fill = np.maximum(y1, y2)  # 取较大值的区域填充
plt.fill_between(x, y_fill, 10, where=(y_fill < 10), color='gray', alpha=0.3)

# 坐标轴
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.legend()
plt.grid()
plt.title("Group Solution Set of Linear Inequalities")
plt.show()
```
